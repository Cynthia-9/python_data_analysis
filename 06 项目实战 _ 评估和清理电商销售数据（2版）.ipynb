# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 评估数据

数据的评估分为结构和内容两部分  
数据结构，即整齐度，数据是否符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”这三个标准；  
数据内容，即干净度，数据是否存在缺失值，重复值，不一致值以及无效值等问题。

### 评估数据整齐度

```python
original_data.sample(10)
```

数据符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”这三个标准，无需对数据整齐度进行调整。

### 评估数据干净度

```python
original_data.info()
```

`Description`产品名称和`CustomerID`顾客id存在缺失值，下面具体看一下缺失值是否影响分析目标；  
此外，`InvoiceDate`的数据类型应为datetime日期，`CustomerID`数据类型为object字符串,应该进行数据类型转换。

#### 缺失数据

`Description`产品名称缺失1454行数据。

```python
original_data[original_data['Description'].isnull()].sample(10)
```

随机取10行`Description`缺失的数据观察得出假设，当`Description`为Nan时，`Unitprice`均为0，下面验证猜想。

```python
original_data[(original_data['Description'].isnull()) & (original_data['UnitPrice'] != 0)]
```

猜想正确，筛选出来结果数量为0条，说明缺失`Description`值的数据，同时也不具备有效的`UnitPrice`值。

`Description`表示产品名称，`UnitPrice`表示产品单价，都是进行后续商品交易分析的重要变量。如果它们同时缺失/无效，我们认为数据无法提供有效含义，因此这些后续可以被删除。

`CustomerID`顾客id缺失135080行数据。

```python
original_data[original_data['CustomerID'].isnull()].sample(10)
```

```python
original_data[(original_data['Description'].isnull()) & (original_data['CustomerID'].isnull())]
```

```python
original_data[(original_data['CustomerID'].isnull()) & (original_data.duplicated())].info()
```

`Description`和`CustomerID`同时为Nan的数据有1454行，执行删除`Description`缺失数据时，均会被删除，无需再次考虑。  
存在`CustomerID`为Nan时，完全重复的数据，此处暂不考虑，放到*评估重复数据*小节讨论。

`CustomerID`表示客户编号，不是分析畅销商品的必要变量。并且从输出结果来看，有些`CustomerID`缺失的交易数据仍然有效，因此保留此变量为空的观察值。

#### 重复数据

```python
original_data[original_data.duplicated()]
```

根据数据变量的含义来看，虽然`InvoiceNo`、`StockCode`和`CustomerID`都是唯一标识符，但一次交易可能包含多件商品，因此`InvoiceNo`可以存在重复，不同交易可以包含同件商品，因此`StockCode`可以存在重复，顾客可以进行多次交易或下单多个商品，因此`CustomerID`也可以存在重复。但是当`InvoiceNo` `InvoiceDate` `CustomerID` `StockCode` `Quantity`同时重复的情况不应当出现，即一个用户在同一时间，同一张发票上购买了同一产品，但是被分开记录了，所以该情况可能是记录人员重复登录销售信息，这些数据会影响我们分析目的的实现，不应予以考虑，应当删除重复记录。  

`CustomerID`为Nan时，完全重复的数据，仅有43行，包含在所有重复数据里一同删除。


#### 不一致数据

不一致数据可能存在于`Country`变量中，我们要查看是否存在多个不同值指代同一国家的情况。

```python
original_data['Country'].value_counts()
```

从`Country`变量值来看，`"USA"`、`"United States"`均在表示美国，`"United Kingdom"`、`"UK"`、`"U.K."`均在表示英国，因此应该在清洗步骤对这些值进行统一，只保留一个指代值。

#### 失效/无效数据

```python
original_data.describe()
```

从输出结果来看，`Quantity`和`UnitPrice`存在负数，会对后续数值分析产生影响。

因此，我们先筛选出`Quantity`数值为负数的观察值，进一步评估其含义。

```python
original_data[(original_data['Quantity'] < 0)]
```

从筛选结果来看，`Quantity`变量为负数的观察值，`InvoiceNo`似乎均以C开头，表示订单被取消。

但此猜测不一定准确，为了验证猜想，我们增加筛选条件，看是否存在`Quantity`变量为负数且`InvoiceNo`不以C开头的观察值。

```python
original_data[(original_data['Quantity'] < 0) & (~ original_data['InvoiceNo'].str.contains('C'))]
```

以上猜想被证实错误，因为还存在`Quantity`变量为负数且`InvoiceNo`不以C开头的观察值。

但根据以上输出结果，这些筛选出的观察值的`UnitPrice`观察值均为0，因此增加`UnitPrice`的条件进行验证。

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0] != "C") & (original_data["UnitPrice"] != 0)]
```

根据输出结果，猜想得到验证，当`Quantity`变量为负数的时候，观察值满足以下条件之一：
- `InvoiceNo`以C开头，表示订单被取消
- `UnitPrice`为0，说明单价为0英镑

这些交易数据均不是有效成交数据，不贡献销售，不在后续分析范围内，因此我们将在数据清理步骤，将`Quantity`为负数的观察值删除。

接下来筛选出`UnitPrice`为零或
负数的观察值。

```python
original_data[original_data['UnitPrice'] < 0]
```

从输出结果来看，`UnitPrice`为负数的观察值都是坏账调账，不属于实际商品交易数据，因此也在数据清理步骤中也将其删除。

```python
original_data[original_data['UnitPrice'] == 0]
```

在*缺失数据*小节，我们验证了当`Description`为Nan值时，`UnitPrice`也均为0.0，此数据有1454行。  

在本小节，我们验证了当`Quantity`< 0 时，且`InvoiceNo`不以C开头，`UnitPrice`为0.0的数据有1336行。  

我们需要筛选出来除以上条件外的其他`UnitPrice` 为0.0 的观察值。

```python
original_data[(original_data['UnitPrice'] == 0) & (original_data['Description'].notnull()) & (original_data["Quantity"] > 0)]
```

在挖掘畅销产品的过程中，销售单价为零的观察值可能会有一定的意义，但也可能是数据中的异常情况。下面是一些考虑销售单价为零观察值意义的因素：

- 免费赠品或促销活动：销售单价为零的观察值可能代表客户获得了免费赠品或者参与了某种促销活动。这种情况下，这些观察值可能有助于理解营销策略的有效性以及客户的购买行为。

- 数据质量问题：销售单价为零的观察值也可能是由于数据质量问题导致的，例如数据输入错误或者数据缺失。在这种情况下，这些观察值可能会干扰畅销产品的挖掘分析，需要谨慎处理或者排除。

- 异常销售情况：销售单价为零的观察值可能表示了某些异常的销售情况，例如退货、取消订单或者系统错误等。这些情况下，这些观察值可能不代表真正的销售情况，需要谨慎处理。

## 清理数据

根据前面评估部分得到的结论，我们需要进行的数据清理包括：  
- 删除完全重复5268行数据；
- 删除缺失`Description`值的数据；
- 删除`Quantity`为负数的观察值；
- 删除`UnitPrice`为负数的观察值（都是坏账调整）；
- 统一`Country`变量值，`"United States"`转换为`"USA"`，`"United Kingdom"`、`"U.K."`转换为`"UK"`；
- 转换`InvoiceDate`的数据类型为datetime日期，转换`CustomerID`数据类型为object字符串；

为了区分开经过清理的数据和原始的数据，我们创建新的变量`cleaned_data`，让它为`original_data`复制出的副本。我们之后的清理步骤都将被运用在`cleaned_data`上。

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

- 删除完全重复5268行数据

```python
cleaned_data[cleaned_data.duplicated()]
```

```python
cleaned_data.drop_duplicates(inplace = True)
cleaned_data
```

检查是否还存在完全重复数据：

```python
541909-536641
```

```python
cleaned_data.duplicated().sum()
```

- 删除缺失`Description`值的数据

```python
cleaned_data.dropna(subset = 'Description' , inplace = True)
```

检查是否还存在`Description`缺失的数据：

```python
536641-535187
```

```python
cleaned_data["Description"].isnull().sum()
```

- 删除`Quantity`为负数的观察值

```python
cleaned_data = cleaned_data[cleaned_data['Quantity'] >= 0]
```

检查是否还存在`Quantity`为负的数据：

```python
cleaned_data[cleaned_data['Quantity'] < 0]
```

```python
len(cleaned_data[cleaned_data['Quantity'] < 0])
```

- 删除`UnitPrice`为负数的观察值（都是坏账调整）

```python
cleaned_data = cleaned_data[cleaned_data['UnitPrice'] >= 0]
```

检查是否还存在`UnitPrice`为负的数据：

```python
len(cleaned_data[cleaned_data['UnitPrice'] < 0])
```

- 统一`Country`变量值，`"USA"`替换为`"United States"`，`"UK"`、`"U.K."`替换为`"United Kingdom"`

```python
cleaned_data = cleaned_data.replace({'USA':'United States' , 'UK':'United Kingdom' , 'U.K.':'United Kingdom'})
```

检查是否还存在'USA','UK','U.K.':

```python
len(cleaned_data[(cleaned_data['Country'] == 'USA') | (cleaned_data['Country'] == 'UK') | (cleaned_data['Country'] == 'U.K.')])
```

- 转换`InvoiceDate`的数据类型为datetime日期，转换`CustomerID`数据类型为object字符串

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate'])
cleaned_data['InvoiceDate'] 
```

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].astype(str)
cleaned_data['CustomerID']
```

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0,-2)
cleaned_data['CustomerID']
```
